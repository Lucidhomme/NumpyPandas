```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#어제 제공된 3개의 파일을 활용한 [실습]
#- ratings.dat
#- movies.dat
#- users.dat

#1. 각 파일을 pd.read_csv() 로 읽어들인다
#- 헤더 및 구분자 확인요망
ratings_df = pd.read_csv('./data/ratings.dat', sep='::', header = None)
movies_df = pd.read_csv('./data/movies.dat', sep='::', header = None)
users_df = pd.read_csv('./data/users.dat', sep='::', header = None)
```

```python
#2. 데이터 확인 및 구조분석과 각 데이터 프레임에 헤더를 추가
#ratings.dat
#- userId movieId rating timestamp

# print(ratings_df.info())
# print(ratings_df.head())
ratings_df.columns = ['userid','movieid','rating','timestamp']
print(ratings_df)

#movies.data
#- movieId title genres

# print(movies_df.info())
# print(movies_df.head())
movies_df.columns = ['movieid', 'title', 'genres']
print(movies_df)

#users.data(뒤에 두열은 제거)
#- userId gender age

print(users_df.info())
print(users_df.head())
users_df.columns = ['userid','gender','age','A','B']
users_df.drop(['A','B'],axis = 1, inplace = True)
print(users_df)
```

```python
#3. 공통 컬럼을 이용한 데이터 병합
first_df = pd.merge(ratings_df, movies_df, on='movieid' , how='inner')
complete_df = pd.merge(first_df, users_df, on='userid' , how='inner')
complete_df
```

```python
#4. user 평점 갯수 순위 확인
m_cnt = pd.DataFrame({'m_cnt' : complete_df['movieid'].value_counts() })
m_cnt.head()

#내가했던거
#rating_count = complete_df['userid'].value_counts()
#rating_count
```

```python
#7. 평점이 200개 이상인 영화들의 성별 평균 평점
complete_df['new_rating'] = complete_df['rating'] if complete_df['rating'].count() >= 200 else ''
complete_df.groupby("gender").mean()
```

```python
#8. 여성이 가장 좋아하는 영화 순위
women_movie = complete_df.sort_values(by=['rating'], ascending=False)
women_movie[women_movie['gender'] == 'F']
```

```python
#9. 남녀간의 평균 평점 차이
a = complete_df[complete_df['gender'] == 'M']
b = complete_df[complete_df['gender'] == 'F']

arating = a['rating'].mean()
brating = b['rating'].mean()

arating - brating
```

```python
# Comedy영화 중 가장 평점이 낮은 영화의 제목
comedy_movie = complete_df.sort_values(by=['rating'], ascending=True)
comedy_movie[comedy_movie['genres'] == 'Comedy']
```

```python
#5. 각 영화의 성별 평균 평점
#pivot()
display(complete_df.pivot_table('rating',['title','gender'],
                              aggfunc='mean',
                              margins=True,
                              margins_name='gender_rating'))
```

```python
#6. 평점이 220개 이상인 영화만 필터링
#groupby()

print(complete_df.groupby('title').size())
print(complete_df.groupby('title').size().index)
rating_by_title = complete_df.groupby('title').size()
rating220 = rating_by_title.index[rating_by_title >= 220]
rating220
```
