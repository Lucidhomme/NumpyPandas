```python
# 판다스를 이용하여 CSV파일 불러오기
import io
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import warnings

df = pd.read_csv(io.StringIO(uploaded['대한무역투자진흥공사 해외진출 한국기업 디렉토리 DB_20161231.csv'].decode('cp949')))
```

```python
df.info()
```

```python
# 칼럼명 확인
df.columns
```

```python
# 칼럼명 변경
df.columns = ['continent','nation','area','ko_company','eng_company','address','homepage','year','foray','investtype','foreignjv','type1','type2','field','dispatch','localrecruit','dom/exp','mothercom','controled by']
df
```

```python
#### 활용할 수 있는 인사이트 : 진출형태 별 개수, 투자형태 별 개수, 업종 별 개수, 현지채용 비중이 높은 나라, 전체 기업에 대한 지도 표시, 대륙/국가/도시 별 개수
```

```python
# 국가 별 현지채용/파견 확인
a = df.groupby(['nation'])['localrecruit'].sum()
b = df.groupby(['nation'])['dispatch'].sum()
```

```python
Nation_local_ratio = a/(a+b)
Nation_local_ratio
```

```python
# 현지채용 비중이 들어간 DF2 새로 만들기
df2 = pd.DataFrame(Nation_local_ratio, columns = ['ratio'])
df2
```

```python
# null값 데이터 개수 확인
df2.isnull().sum()
```

```python
# null값 데이터 찾기 : 원인 미상
df2.loc[df2['ratio'].isna()]
```

```python
# DF2개를 서로 머지하여 국가 별 ratio를 입력
df2.reset_index(level='nation') # 인덱스 열로 바꾸기
df3 = pd.merge(df, df2, on = 'nation', how = 'left') # Merge
```

```python
# 비중이 null값인 데이터 개수확인
df3.isnull().sum()
```

```python
# 비중이 null값인 데이터 찾기 - 독일, 세르비아, 페루
df3.loc[df3['ratio'].isnull()]
```

```python
#NaN 값 일괄적으로 0으로 기입
df3['ratio'].fillna(0)
```

```python
#그래도 결측값이 대체가 안되는 상태
df3.isnull().sum()
```

```python
# dropna로 제거
df4 = df3.dropna(subset = ['ratio'])
```

```python
# 결측행 제거완료
df4.isnull().sum()
```

```python
# 상위 5개, 하위 5개 국가 출력
top5 = df4.sort_values(by='ratio',ascending=False).groupby('nation').head(5)
bottom5 = df4.sort_values(by='ratio',ascending=True).groupby('nation').head(5)
```
