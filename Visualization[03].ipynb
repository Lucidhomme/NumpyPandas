```python
# 파일 읽기 : 서울지역 초등학교 정보
seoul_ele_df = pd.read_excel('./data/elem_school_folium.xlsx')
seoul_ele_df.head()
```

```python
# Feature selection
# 학교명, 학교급구분, 소재지도로명, 위도, 경도
feature_sel_df = seoul_ele_df.filter(['학교명','학교급구분','소재지도로명주소','위도','경도'])
feature_sel_df.head()
```

```python
# 소재지도로명주소가 서울을 포함하고 학교급구분 초등학교만 추출
# sql like
# query

#인덱스 추출
index_seoul = feature_sel_df['소재지도로명주소'].str.contains('서울특별시').values
index_ele = feature_sel_df['학교급구분'] == '초등학교'
print(len(index_seoul))
print(len(index_ele))

result_df = feature_sel_df.loc[index_seoul & index_ele]
result_df.tail()


```

```python
for name, lat, lng in zip(result_df.index , result_df.위도, result_df.경도) :
    g.CircleMarker([lat, lng], 
                   popup=name, 
                   icon=g.Icon(color='red'),
                  radius=10,
                  fill=True,
                  fill_color='coral',
                  fill_opacity=0.3).add_to(seoul_map)

seoul_map
```

```python
# 경계정보를 가진 json 파일 읽어보자
geo_data = json.load(open('./data/경기도행정구역경계.json', encoding = 'UTF-8'))
geo_data
```

```python
pop_df.columns = pop_df.columns.map(str)

gyeonggi_map = g.Map(location=[37.55, 126.98],
                    zoom_start=18,
                    tiles='Stamen Terrain')

# 구분된 경계를 표시할 때 사용하는 객체
year = '2017'
g.Choropleth(geo_data = geo_data , 
            data = pop_df['2017'],
            columns = [pop_df.index, pop_df[year]],
            key_on = 'feature.properties.name').add_to(gyeonggi_map)

gyeonggi_map
```
